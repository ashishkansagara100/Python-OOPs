```python
#. Animal Dog speak override
class Animal:
    def speak(self):
        print("Some generic animal sound")

class Dog(Animal):
    def speak(self):
        print("Bark!")

d = Dog()
d.speak()
```

```python
#2. Abstract Shape Circle Rectangle
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        print("Circle area:", 3.14*self.r*self.r)

class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        print("Rectangle area:", self.l*self.b)
        
c = Circle(5)
c.area()

r = Rectangle(4,5)
r.area()
```

```python
#3. Multi level inheritance Vehicle Car ElectricCar
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def show(self):
        print("Vehicle type:", self.type)

class ElectricCar(Car):
    def __init__(self, type, battery):
        super().__init__(type)
        self.battery = battery

    def show(self):
        print("Type:", self.type)
        print("Battery:", self.battery)

e = ElectricCar("Car", "5000mAh")
e.show()
```

```python
#4. Bird polymorphism
class Bird:
    def fly(self):
        print("Some birds can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

s = Sparrow()
p = Penguin()

s.fly()
p.fly()
```

```python
#5. BankAccount encapsulation
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amt):
        self.__balance += amt

    def withdraw(self, amt):
        self.__balance -= amt

    def show(self):
        print("Balance:", self.__balance)

b = BankAccount()
b.deposit(1000)
b.withdraw(200)
b.show()
```

```python
#6. Instrument runtime polymorphism
class Instrument:
    def play(self):
        print("Playing instrument")

class Guitar(Instrument):
    def play(self):
        print("Playing Guitar")

class Piano(Instrument):
    def play(self):
        print("Playing Piano")

g = Guitar()
p = Piano()

g.play()
p.play()
```

```python
#7. Class method and static method
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        print("Add:", a+b)

    @staticmethod
    def subtract_numbers(a, b):
        print("Subtract:", a-b)

MathOperations.add_numbers(5,3)
MathOperations.subtract_numbers(10,4)
```

```python
#9. Fraction str method
class Fraction:
    def __init__(self, n, d):
        self.n = n
        self.d = d

    def __str__(self):
        return f"{self.n}/{self.d}"

f = Fraction(3,4)
print(f)
```

```python
#10. Vector operator overloading
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x+other.x, self.y+other.y)

    def __str__(self):
        return f"({self.x},{self.y})"

v1 = Vector(2,3)
v2 = Vector(3,4)

print(v1 + v2)
```

```python
#12. Student average grade
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        print("Average:", sum(self.grades)/len(self.grades))

s = Student("A", [80,90,85])
s.average_grade()
```

```python
#14. Employee Manager salary
class Employee:
    def calculate_salary(self, hours, rate):
        return hours * rate

class Manager(Employee):
    def calculate_salary(self, hours, rate):
        return super().calculate_salary(hours, rate) + 5000

e = Employee()
m = Manager()

print("Employee salary:", e.calculate_salary(8,500))
print("Manager salary:", m.calculate_salary(8,500))
```

```python
#15. Product total price
class Product:
    def __init__(self, name, price, qty):
        self.name = name
        self.price = price
        self.qty = qty

    def total_price(self):
        print("Total:", self.price*self.qty)

p = Product("Book", 100, 5)
p.total_price()
```

```python
#16. Animal sound abstract
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

c = Cow()
s = Sheep()

c.sound()
s.sound()
```

```python
#17. Book info
class Book:
    def __init__(self, t, a, y):
        self.title = t
        self.author = a
        self.year = y

    def get_book_info(self):
        return self.title + " " + self.author + " " + str(self.year)

b = Book("ABC","XYZ",2020)
print(b.get_book_info())
```

```python
#18. House Mansion
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, rooms):
        super().__init__(address, price)
        self.rooms = rooms

m = Mansion("Mumbai", 5000000, 10)
print(m.address, m.price, m.rooms)
```
